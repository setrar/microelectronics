# Impedances

## Frequency 

The **same signal frequency** can be represented in both **Hertz (Hz)** and **radians per second (rad/s)**, but they are different units and convey slightly different information. Here's how they relate to each other:

---

### **1. Hertz (Hz):**
   - **Definition:** Hertz is the unit of frequency in cycles per second.
   - **Meaning:** It represents how many complete cycles (oscillations) of a signal occur in one second.
   - **Example:** A signal with a frequency of 50 Hz completes 50 cycles per second.

---

### **2. Radians per Second (rad/s):**
   - **Definition:** Radians per second is the unit of angular frequency.
   - **Meaning:** It represents the rate of change of the phase angle of the signal in radians per second.
   - **Example:** A signal with an angular frequency of 100π rad/s oscillates at a rate of 100π radians per second.

---

### **Relationship Between Hz and rad/s:**
The two units are related by the following formula:

$\omega = 2\pi f$

Where:
- $\omega$ = angular frequency in **rad/s**
- $f$ = frequency in **Hz**
- $2\pi$ = conversion factor (since one full cycle corresponds to $2\pi$ radians)

---

### **Example Conversion:**
If a signal has a frequency of $f = 50 \, \text{Hz}$, its angular frequency $\omega$ is:

$\omega = 2\pi \times 50 = 100\pi \, \text{rad/s}$

Similarly, if a signal has an angular frequency of $\omega = 200\pi \, \text{rad/s}$, its frequency in Hz is:

$f = \frac{\omega}{2\pi} = \frac{200\pi}{2\pi} = 100 \, \text{Hz}$

---

### **Key Differences:**
| **Aspect**            | **Hertz (Hz)**                     | **Radians per Second (rad/s)**       |
|------------------------|-------------------------------------|--------------------------------------|
| **Unit Type**          | Frequency (cycles per second)      | Angular frequency (radians per second) |
| **Physical Meaning**   | Number of cycles per second        | Rate of change of phase angle        |
| **Formula**            | $f$                             | $\omega = 2\pi f$                  |
| **Usage**              | General signal analysis            | Used in equations involving phase or angular motion (e.g., AC circuits, control systems) |

---

### **When to Use Which:**
- **Hertz (Hz):** Used when describing the frequency of signals in terms of cycles per second (e.g., power line frequency, audio signals).
- **Radians per Second (rad/s):** Used in mathematical analysis, especially in equations involving phase angles, such as in AC circuit analysis, control systems, and signal processing.

---

### **Summary:**
- The same signal frequency can be represented in **Hz** (cycles per second) or **rad/s** (radians per second).
- They are related by the formula $\omega = 2\pi f$.
- **Hz** is more intuitive for practical applications, while **rad/s** is more convenient for mathematical and theoretical analysis.

## Visualizing **radians per second (rad/s)** 

Visualizing **radians per second (rad/s)** can be a bit abstract at first, but it becomes clearer when you connect it to the motion of a rotating object or the oscillation of a wave. Let me break it down for you:

---

### **1. What is a Radian?**
   - A **radian** is a unit of angular measure.
   - It is defined as the angle subtended at the center of a circle by an arc whose length is equal to the radius of the circle.
   - In simpler terms: If you take a circle and "unroll" its radius along the circumference, the angle created at the center is **1 radian**.
   - A full circle ($360^\circ$) corresponds to $2\pi$ radians (since the circumference of a circle is $2\pi r$, where $r$ is the radius).

---

### **2. What Does Radian/Second (rad/s) Represent?**
   - **Radian per second (rad/s)** is a measure of **angular velocity**.
   - It tells you how fast an object is rotating or how quickly an angle is changing over time.
   - For example:
     - If an object rotates at **1 rad/s**, it means the angle changes by **1 radian every second**.
     - If it rotates at **2π rad/s**, it completes **one full rotation (360°)** every second.

---

### **3. Visualizing Rad/s:**
Here are some ways to visualize **rad/s**:

#### **a. Rotating Wheel or Fan:**
   - Imagine a wheel or fan spinning.
   - If it rotates at **1 rad/s**, it means the wheel turns through an angle of **1 radian (≈57.3°)** every second.
   - If it rotates at **2π rad/s**, it completes **one full rotation (360°)** every second.

#### **b. Pendulum or Oscillating Wave:**
   - Think of a pendulum swinging back and forth or a sine wave oscillating.
   - The angular frequency ($\omega$) in **rad/s** tells you how quickly the pendulum or wave oscillates.
   - For example:
     - A pendulum with $\omega = 2 \, \text{rad/s}$ oscillates at a rate of **2 radians per second**.
     - A sine wave with $\omega = 100\pi \, \text{rad/s}$ oscillates much faster.

#### **c. Clock Hands:**
   - The second hand of a clock rotates at **$\frac{\pi}{30} \, \text{rad/s}$**:
     - It completes one full rotation ($2\pi$ radians) in **60 seconds**.
     - So, $\omega = \frac{2\pi}{60} = \frac{\pi}{30} \, \text{rad/s}$.

---

### **4. Relating Rad/s to Hz:**
   - **Hertz (Hz)** measures frequency in **cycles per second**.
   - **Rad/s** measures angular frequency in **radians per second**.
   - The two are related by the formula:
$\omega = 2\pi f$
     where:
     - $\omega$ = angular frequency in **rad/s**
     - $f$ = frequency in **Hz**
   - For example:
     - A signal with $f = 1 \, \text{Hz}$ has $\omega = 2\pi \, \text{rad/s}$.

---

### **5. Practical Example:**
   - Suppose a wheel rotates at **10 rad/s**.
   - This means:
     - The wheel turns through **10 radians every second**.
     - Since $2\pi \approx 6.28$ radians is one full rotation, the wheel completes:
$\frac{10}{2\pi} \approx 1.59 \, \text{rotations per second}.$
     - In terms of frequency, this is approximately **1.59 Hz**.

---

### **6. Summary:**
- **Rad/s** measures how quickly an angle changes over time.
- It can be visualized as the speed of rotation (e.g., a spinning wheel) or the rate of oscillation (e.g., a pendulum or wave).
- **1 rad/s** means the angle changes by **1 radian every second**.
- To convert to **Hz**, use $\omega = 2\pi f$.
