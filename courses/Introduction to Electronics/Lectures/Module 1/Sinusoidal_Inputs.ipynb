# **sinusoidal inputs**

## **Steady-state sinusoidal inputs**

**Steady-state sinusoidal inputs** refer to sinusoidal signals (e.g., sine or cosine waves) that are applied to a system and have been present for a sufficiently long time so that any transient effects (initial disturbances or responses) have decayed to zero. At this point, the system's response is entirely characterized by the sinusoidal input, and the output is also a sinusoidal signal with the same frequency but potentially different amplitude and phase.

---

### **1. Key Characteristics of Steady-State Sinusoidal Inputs**
   - **Sinusoidal Waveform:** The input signal is of the form:
$x(t) = A \sin(\omega t + \phi)$
     where:
     - $A$ = amplitude
     - $\omega$ = angular frequency (in rad/s)
     - $\phi$ = phase angle (in radians)
   - **Steady-State:** The system has reached a stable, long-term response to the input, and transient effects have disappeared.
   - **Same Frequency:** The output signal has the same frequency as the input signal but may have a different amplitude and phase.

---

### **2. Why Sinusoidal Inputs are Important**
   - Sinusoidal signals are fundamental in analyzing **linear time-invariant (LTI) systems** because:
     - They are **eigenfunctions** of LTI systems, meaning the output is also a sinusoid with the same frequency.
     - They allow us to study system behavior in the **frequency domain**.
   - Sinusoidal inputs are used to characterize systems in terms of their **frequency response** (amplitude and phase changes at different frequencies).

---

### **3. Steady-State Response to Sinusoidal Inputs**
   - For an LTI system with a sinusoidal input, the steady-state output is:
$y(t) = B \sin(\omega t + \phi + \theta)$
     where:
     - $B$ = amplitude of the output (depends on the system's gain at frequency $\omega$)
     - $\theta$ = phase shift introduced by the system
   - The relationship between the input and output is described by the system's **frequency response**:
$H(j\omega) = \frac{Y(j\omega)}{X(j\omega)}$
     where:
     - $H(j\omega)$ = frequency response (complex function)
     - $Y(j\omega)$ = Fourier Transform of the output
     - $X(j\omega)$ = Fourier Transform of the input

---

### **4. Frequency Response and Impedance**
   - In electrical circuits, the **frequency response** is often related to **impedance**:
     - For a circuit with impedance $Z(j\omega)$, the output voltage ($V_{\text{out}}$) for a sinusoidal input voltage ($V_{\text{in}}$) is:
$V_{\text{out}}(j\omega) = H(j\omega) \cdot V_{\text{in}}(j\omega)$
     - The frequency response $H(j\omega)$ depends on the circuit's impedance and topology.

---

### **5. Applications of Steady-State Sinusoidal Inputs**
   - **AC Circuit Analysis:** Used to analyze circuits with alternating current (AC) signals.
   - **Filter Design:** Used to design and analyze filters (e.g., low-pass, high-pass, band-pass).
   - **Control Systems:** Used to study the frequency response of control systems and design controllers.
   - **Signal Processing:** Used to analyze and process signals in communication systems, audio systems, and more.
   - **Impedance Matching:** Used to match impedances in RF and audio systems for maximum power transfer.

---

### **6. Example: RC Circuit**
   - Consider an **RC circuit** (resistor and capacitor in series) with a sinusoidal input voltage:
$V_{\text{in}}(t) = A \sin(\omega t)$
   - The steady-state output voltage across the capacitor is:
$V_{\text{out}}(t) = \frac{A}{\sqrt{1 + (\omega RC)^2}} \sin(\omega t + \theta)$
     where:
     - $\theta = -\tan^{-1}(\omega RC)$ (phase shift)
   - The frequency response of the RC circuit is:
$H(j\omega) = \frac{1}{1 + j\omega RC}$

---

### **7. Summary**
   - **Steady-state sinusoidal inputs** are sinusoidal signals applied to a system after transient effects have decayed.
   - The system's response is also sinusoidal, with the same frequency but potentially different amplitude and phase.
   - These inputs are fundamental for analyzing **LTI systems** in the frequency domain and are widely used in **AC circuits**, **filters**, **control systems**, and **signal processing**.
